```python
import requests
from bs4 import BeautifulSoup

html='''
<html>
<body>
    <div class="product-list">
        <div class="product">
            <h3 class="name">스마트폰</h3>
            <p class="price">1,000,000원</p>
            <span class="stock">재고: 5개</span>
        </div>
        <div class="product">
            <h3 class="name">노트북</h3>
            <p class="price">2,000,000원</p>
            <span class="stock">재고: 3개</span>
        </div>
    </div>
</body>
</html>
'''

soup = BeautifulSoup(html,"html.parser")
products =soup.select("div.product")

for product in products:
    name = product.select_one('h3.name').text
    price = product.select_one('p.price').text
    stock = product.select_one('span.stock').text
    print(f"상품명: {name}, 가격: {price},정보:{stock}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

html = '''
<html>
<body>
    <table class="student-table">
        <tr>
            <th>학생명</th>
            <th>학년</th>
            <th>전공</th>
        </tr>
        <tr>
            <td>이영희</td>
            <td>3</td>
            <td>컴퓨터 공학</td>
        </tr>
        <tr>
            <td>박민수</td>
            <td>2</td>
            <td>디지털 디자인</td>
        </tr>
        <tr>
            <td>최지우</td>
            <td>1</td>
            <td>경영학</td>
        </tr>
    </table>
</body>
</html>
'''
soup = BeautifulSoup(html,"html.parser")
table = soup.select_one(".student-table")

data = []
for row in table.select('tr')[1:]:
    cols = row.select("td")
    data.append([col.text for col in cols])
print(data)
df = pd.DataFrame(data,columns= [th.text for th in table.select('tr')[0].select('th')])
#df

with open("student.csv","w",newline="",encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow([th.text for th in table.select('tr')[0].select('th')])
    writer.writerows(data)
    
    
```
